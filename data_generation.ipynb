```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
# Generate 4 blobs of data
centers = [[1,1],[-1,1],[1,-1],[-1,-1]]
X,y = make_blobs(n_samples=1000, n_features=2, centers=centers, cluster_std=0.55)
X1 = X.T[0]
X2 = X.T[1]
plt.scatter(X1,X2)
```

```python
# Save data to CSV
df = pd.DataFrame(X)
df.to_csv('cluster_data.csv', index=False, index_label=False)
```
